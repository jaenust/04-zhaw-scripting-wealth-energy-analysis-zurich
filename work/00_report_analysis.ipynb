# 00_report_analysis

## Zusammenfassung der Analyse

Dieses Notebook untersucht den Zusammenhang zwischen Wohlstandsindikatoren und Energieverbrauch in Zürcher Gemeinden.

### Datenquellen
- **Energie**: Gesamt- und Haushaltsstromverbrauch pro Kopf (2024)
- **Wohlstand**: Einkommens- und Vermögenskennzahlen (Median, Durchschnitt, pro Kopf) aus 2022
- **Geografie**: Gemeindeboundaries des Kantons Zürich

### Durchgeführte Analysen

1. **Umfassende Korrelationsanalyse**
   - Berechnung von Pearson- und Spearman-Korrelationen für alle Kombinationen
   - Energietypen: Haushaltsstrom, Gesamtstrom
   - Wohlstandskennzahlen: Einkommen (Median/Durchschnitt/pro Kopf), Vermögen (Median/Durchschnitt/pro Kopf)
   - Ergebnisse gespeichert in `comprehensive_correlations.csv`

2. **Visualisierungen**
   - Scatterplots mit Wohlstandsmetriken auf der Y-Achse und Energieverbrauch auf der X-Achse
   - Korrelations-Heatmap zur gleichzeitigen Darstellung aller Zusammenhänge
   - Choroplethenkarten zur räumlichen Darstellung der Kennzahlen
   - Alle Grafiken gespeichert unter `outputs/figures/` und `outputs/maps/`

3. **Ausreißeranalyse**
   - Identifikation von Gemeinden mit ungewöhnlichen Energieverbrauchsmustern
   - Basierend auf Regressionsresiduen und Z-Scores

### Zentrale Erkenntnisse
Die Korrelationstabellen und Grafiken zeigen auf:
- Welche Wohlstandsmetriken am stärksten mit dem Energieverbrauch korrelieren
- Ob Haushaltsstrom oder Gesamtstrom stärkere Zusammenhänge aufweist
- Räumliche Muster in der Beziehung zwischen Wohlstand und Energieverbrauch
